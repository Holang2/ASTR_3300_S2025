```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import scipy.optimize as ot
import numdifftools as nd
import corner

# Part 1
# Loading the sample data
data = np.load('C:/Users/JoeK33p/Documents/GitHub/ASTR_3300_S2025/coursework/homeworks/hw_data/hw3_data_1.npy')

# reading in the measurements and the uncertainties
meas = data[:,0]
uncert = data[:,1]

# Calculating the sample mean and standard deviantion
mean = np.sum(meas/uncert**2)/np.sum(1/uncert**2)
sigma = 1/np.sqrt(np.sum(1/uncert**2))

print("The mean of the sample data is ",mean)
print("The standard deviation of the sample data is ",sigma)

# Part 2
# Defining the ln likelihood function for gauss
def lnl_g(mu, data):
    meas = data[:,0]
    uncert = data[:,1]
    ln_like = -0.5*np.sum((np.log(2*np.pi*uncert**2)+(meas-mu)**2) / uncert**2)
    return ln_like

# Finding the mean best fit of the gauss ln likelihood 
avg = np.mean(meas)
out_ln_g = lambda mu: -lnl_g(mu, data)
gauss = ot.minimize(out_ln_g, avg)
fit_g = gauss.x[0]
print("The mean of the best fit gauss is", fit_g)

# Part 3
# Numerical technique for finding fisher error
st_dev = np.std(meas)
mu_data = np.mean(meas)
xgrid = np.linspace(0,7,100)
distg = stats.norm(mu_data, st_dev)
L = distg.pdf(xgrid)
sigma_mu = np.diff(np.log(L),n=2)
sigma_mu /= (xgrid[1]-xgrid[0])**2
sigma_mu *= -1
sgma_mu = 1/np.sqrt(sigma_mu)


print("The fisher uncertainty estimate is ", sigma_mu[0])

# Part 4
# Boot strapping
def bootstrap(data, n_s=1000):
    n_data = len(data)
    b_sam = []
    for _ in range(n_s):
        ind = np.random.choice(n_data, n_data, replace=True)
        b_sam.append(data[ind])
    b_sam = np.array(b_sam)
    return b_sam

bs_s = bootstrap(data)

# Part 5
# Finding the means of bootstrap and comparing them
bs_avg = [np.mean(s[:, 0]) for s in bs_s]

plt.hist(bs_avg, bins = 35, density = True, alpha = 0.5, color = 'orange', label = "Bootstrap Means")
x = np.linspace(mean - 3 * sigma, mean + 3 * sigma, 100)
plt.plot(x, stats.norm.pdf(x, mean, sigma), color = 'green', label = "Analytic Gaussian")
plt.axvline(mean, color = 'black', alpha = 0.5, ls = 'dashed')
plt.xlabel('Mean')
plt.ylabel('Probability of Data')
plt.title('Bootstrap vs Analytic Gaussian')
plt.legend()
plt.show()

# Part 6
# Calculting the best fit Laplacian likelihood 
def ln_l(params, data):
    mu, delta = params
    meas = data[:,0]
    ln_l_val = -np.sum(np.log(2*delta)+np.abs(meas-mu)/delta)
    return ln_l_val

lap = ot.minimize(lambda params: -ln_l(params, data), [mu_data, st_dev])
fit_mu_l, fit_delta_l = lap.x

print("The best fit Laplacian mean is ",fit_mu_l)
print("The best fit Laplacian scale is ",fit_delta_l)


# Part 7
# Calculating the AIC 
def aic(ln_mu, k, data):
    meas = data[:,0]
    N = len(meas)
    aic = -2*ln_mu + 2*k + (2*k*(k-1))/(N-k-1)
    return aic

aic_gauss = aic(lnl_g(fit_g, data), 1, data)
l_params = [fit_mu_l, fit_delta_l]
aic_params = np.array(l_params)
aic_lap = aic(ln_l(l_params, data), 2, data)

print('The aic of the gaussian is ', aic_gauss)
print('The aic of the laplacian is ', aic_lap)

if aic_gauss < aic_lap:
    print('The Gaussian model is favored.')
else:
    print('The Laplacian model is favored.')
    
# Part 8
# Finding the bootstrap of the laplacian fit
# I keep getting a runtime error for my log function and I dont know why it is not working now
bs_l_params = []
for s in bs_s:
    bs_l = ot.minimize(lambda params: -ln_l(params, s), [np.mean(s[:,0]), np.std(s[:,0])])
    bs_l_params.append(bs_l.x)
    
bs_l_params = np.array(bs_l_params)

fig = corner.corner(bs_l_params, labels = ['mu', 'delta'], truths = [fit_mu_l, fit_delta_l], quantiles = [0.68,0.95], title_fmt = '.4f')
plt.show()

# Part 9
# Calculating the hessian lapce

def h_l(params, data):
    mu, delta = params
    h_f = nd.Hessian(lambda params: -ln_l(params, data))
    return h_f([mu, delta])

h_l_m = h_l(l_params, data)
f_c_l = np.linalg.inv(h_l_m)
f_u_l = np.sqrt(np.diag(f_c_l))

print('The Fisher uncertainty for the best fit mu of the Laplacian is ', f_u_l[0])
print('The Fisher uncertainty for the best fit delta of the Laplacian is ', f_u_l[1])


```
